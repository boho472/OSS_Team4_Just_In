```python
import os
import re
import json

label_root = "src/data/KITTI/tracking/training/label_02"
save_path = "src/data/new_data"

num_of_labels = os.listdir(label_root)

for i in range(len(num_of_labels)):
  path = os.path.join(label_root,num_of_labels[i])
  input_lines = []
  with open(path) as f:
    for line in f.readlines():
      line_split = re.split(" ", line.strip())
      if line_split[2] == "Pedestrian":
        input_lines.append(line)

  os.path.join(save_path, str(i).zfill(4)+'.json')
  with open(os.path.join(save_path, str(i).zfill(4)+'.json'), "w") as f:
      json.dump(input_lines, f, indent=2)

  with open(os.path.join(save_path, num_of_labels[i]), "w") as f:
      for lines in input_lines:
          f.write(lines)

print("process done.")
```
